```python
import yt_dlp

# YouTube 동영상 URL
video_url = 'https://www.youtube.com/watch?v=YzTQLsD5DHg'

# 다운로드 및 변환 옵션 설정
ydl_opts = {
    'format': 'bestaudio/best',  # 최고 오디오 품질로 다운로드
    'postprocessors': [{  # mp3로 변환
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'outtmpl': 'youtube_output',  # 출력 파일명 설정
}

# yt-dlp 실행
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([video_url])
```

# Whisper 모델이란?
OpenAI에서 개발한 자동 음성 인식(ASR, Automatic Speech Recognition) 모델. 이 모델은 다양한 언어와 다양한 환경에서 녹음된 음성을 텍스트로 변환하는데 탁월한 성능을 보여줌. 특히, Whisper는 여러 언어와 사투리, 잡음이 섞인 음성까지도 인식할 수 있도록 광범위한 데이터로 훈련되어 있음.

* ### 특징:

**다양한 언어 지원:** Whisper는 한 언어뿐만 아니라 여러 언어의 음성을 인식\
**잡음 처리:** 주변 소음이 있는 환경에서도 음성을 잘 인식할 수 있도록 훈련\
**다양한 모델 크기:** Whisper는 tiny, base, small, medium, large 등 다양한 크기의 모델이 있음. 크기가 클수록 더 높은 인식 정확도를 보이지만, 처리 속도와 메모리 사용량이 증가.

```python
import whisper

#  작은 크기의 Whisper 모델 불러오기
model = whisper.load_model("small")
"""
이유: 속도 향상, 메모리 사용량 감소, 실용성 향상
"""

# 지정된 오디오 파일을 텍스트로 변환
result = model.transcribe("/content/youtube_output.mp3")
result
```

```python
result['text']
```

```python
for temp in result['segments']:
    print(f'[{temp["start"]:.2f} --> {temp["end"]:.2f}]  {temp["text"]}')
```

```python
#음성추출(wavs 디렉토리 생성필수)

"""
ffmpeg를 사용하여 원본 오디오 파일(youtube_output.mp3)에서
특정 구간(start~end)만 추출하여 새로운 wav 파일로 저장

-ss는 시작 시간, -to는 끝나는 시간을 지정,
-hide_banner와 -loglevel error는 ffmpeg 출력 메시지를
최소화하기 위한 옵션
"""
for index,temp in enumerate(result['segments']): # result['segments']의 각 세그먼트를 순회하며 인덱스와 내용을 가져옴
    !ffmpeg -y -i youtube_output.mp3 -ss {temp["start"]} -to {temp["end"]} -hide_banner -loglevel error wavs/audio{index+1}.wav
```

```python
# 텍스트 데이터를 파일로 저장
with open('textdata.txt', 'w', encoding='utf-8') as f:  # 'textdata.txt' 파일을 쓰기 모드로 열고, UTF-8 인코딩을 사용.
    for index, temp in enumerate(result['segments']):  # result['segments']의 각 세그먼트를 순회하며 인덱스와 내용을 가져옴
        # 각 세그먼트의 인덱스와 텍스트를 파일에 쓰기. 'audio{index+1}|텍스트|텍스트' 형식으로 저장
        f.write(f'audio{index+1}|{temp["text"].strip()}|{temp["text"].strip()}\n')
```

* TTS 클래스의 인스턴스를 생성
* model_name은 사용할 TTS 모델을 지정. 이 모델은 다국어를 지원하는 "xtts_v2" 모델

* gpu=True로 설정하여 GPU를 사용해 더 빠르게 음성 변환을 수행. (GPU가 없는 경우 False로 설정.)

```python
import scipy# 음성 처리와 과학 계산을 위한 scipy 라이브러리
from TTS.api import TTS # TTS(TTS: Text-to-Speech) 라이브러리에서 TTS 클래스 임포트

# TTS 모델 불러오기
tts = TTS(model_name="tts_models/multilingual/multi-dataset/xtts_v2",gpu=True)
```

```python
# 속도를 조절하여 음성 합성
tts.tts_to_file(
    text='안녕하세요. 여러분, 저는 서인국 입니다.',
    file_path='./output.wav',
    speaker_wav='/content/wavs/audio15.wav',
    language='ko',
    speed=1.0,  # 속도를 조절 (1.0은 기본 속도, 0.8은 느리게, 1.2는 빠르게)
    # pitch argument removed as it is not supported
)
```
